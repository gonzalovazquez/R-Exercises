```r
dim(tophitters2001) # dimension
tophit <- tophitters2001[1:25, ]
ggplot(tophit, aes(x=avg, y=name)) + geom_point()
```

```r
ggplot(tophit, aes(x=avg, y=name)) + geom_point(aes(color = g)) # add another dimension via color scale
```

```r
ggplot(tophit, aes(x=avg, y=name)) + geom_point(aes(alpha = g)) # via opacity
```

```r
ggplot(tophit, aes(x=avg, y=name)) + geom_point(aes(size = g)) # via size
```

```r
table(tophit$lg)
ggplot(tophit, aes(x=avg, y=name)) + geom_point(aes(shape = factor(lg))) # lg has discrete values
```

```r
ggplot(tophit, aes(x=avg, y=reorder(name, avg))) + geom_point() # re-order according to avg

#improvement
```

```r
ggplot(tophit, aes(x=avg, y=reorder(name, avg))) + geom_point(size=3) + theme_bw() + theme(panel.grid.major.x = element_blank(), panel.grid.minor.x = element_blank(), panel.grid.major.y = element_line(colour="grey60", linetype="dashed"))
# use a larger dot, white background, horizontal dashed grid lines 
```

```r
ggplot(tophit, aes(x=reorder(name, avg), y=avg)) + geom_point(size=3) + theme_bw() + theme(axis.text.x = element_text(angle=60, hjust=1), 
              panel.grid.major.y = element_blank(),
              panel.grid.minor.y = element_blank(),
              panel.grid.major.x = element_line(colour="grey60", linetype="dashed"))
# plot avg against players
```

```r
# example of heatmap

data(presidents)
```

```r
class(presidents) 
# this is time-series data
```

```r
plot(presidents) # this is the default way to plot this type of data
```

```r
# let's say we don't like this and we want to plot is as a heat map
# ts objects use these functions

time(presidents) # time at which the ts was sampled
cycle(presidents) # positions in the cycle of each observation

pres_rating <- data.frame(
        rating  = as.numeric(presidents),
        year    = as.numeric(floor(time(presidents))),
        quarter = as.numeric(cycle(presidents))
)
pres_rating
```

```r
# data now have the so called "long" format
# values were taken in a row-wise fashion
p <- ggplot(pres_rating, aes(x=year, y=quarter, fill=rating))
p+geom_tile()
```

```r
# some changes
tiff("heatmap.tiff")
p+geom_tile() + scale_x_continuous(breaks=seq(1940,1976, by=4)) + scale_y_reverse()
dev.off()
```

```r
# remove the gray background
p+geom_tile() + scale_x_continuous(breaks=seq(1940,1976, by=4)) + scale_y_reverse() + theme_bw()
```

```r
# also it can be done using heatmap "core" function
heatmap(matrix(as.numeric(presidents),nrow=4,ncol=30))
# check the syntax of the function from the documentation
```
