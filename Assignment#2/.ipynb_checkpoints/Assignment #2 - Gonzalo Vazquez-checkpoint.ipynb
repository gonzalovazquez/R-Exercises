```r
# Load iris dataset
library(datasets)
data(iris)
summary(iris)
```

```r
# Without ggplot
```

```r
# Distribution of sepal length
hist(iris$Sepal.Length,main="Distribution of Sepal Length", xlab="Sepal Length")
```

```r
# Visualizing the distribution of sepal width against grouping based on species
boxplot(Sepal.Width ~ Species, data=iris, xlab="Species",ylab="Sepal Width")
```

```r
# Show the relationship of sepal length and petal length
plot(iris$Sepal.Length, iris$Petal.Length, main="Relationship of Sepal Length and Petal Length", 
     col="grey", ylab = 'Petal Length', xlab = "Sepal Length")
```

```r
# Present the pairwise relationship of all four continous variables
plot(iris)
```

```r
# Graph the frequency of occurrences for the three groups
heatmap(matrix(iris$Species))
```

```r
# Install GGPLOT2
install.packages("ggplot2", repos = "http://cran.us.r-project.org")
library(ggplot2)
```

```r
# Distribution of Sepal Length
ggplot(iris, aes(x=Sepal.Length)) + geom_histogram(binwidth=.2)
```

```r
# Visualizing the distribution of sepal width against grouping based on species
ggplot(iris, aes(x=Species, y=Sepal.Width)) + geom_boxplot()
```

```r
# Show the relationship of sepal length and petal length
ggplot(iris, aes(x=Sepal.Length, y=Petal.Length)) + geom_jitter()
```

```r
# Present the pairwise relationship of all four continous variables
#ggplot(iris) + geom_jitter()
```

```r
# Graph the frequency of occurrences for the three groups
ggplot(iris, aes(c(Species))) + geom_freqpoly(binwidth=.2)
```
